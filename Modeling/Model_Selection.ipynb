```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(solver = 'newton-cg', multi_class='multinomial') 
logreg.fit(X1_train,Y1_train)
logreg_train_acc = logreg.score(X1_train,Y1_train)
logreg_test_acc = logreg.score(X1_test,Y1_test)
Xnew = [[1,3,0,2,3,4,0,0]]
ynew_result = logreg.predict(Xnew)

ynew = logreg.predict_proba(Xnew)

print(max(max(ynew)))
print(ynew)
print(ynew_result)
```

```python
print ('logreg training acuracy= ',logreg_train_acc)
print('logreg test accuracy= ',logreg_test_acc)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.tree import DecisionTreeClassifier
knn = KNeighborsClassifier(n_neighbors = 5)                  # instantiate
knn.fit(X1_train, Y1_train)                                    # fit
acc_knn = knn.score(X1_test, Y1_test)                            # predict + evaluate

print('K-Nearest Neighbors labeling accuracy:', str(round(acc_knn*100,2)),'%')                                
```
